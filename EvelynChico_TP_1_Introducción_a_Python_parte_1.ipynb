### **Ejercicio 29**

Construir un programa que lea un número natural **N** y calcule la suma de los primeros **N** números pares.
-  Si N = 2 -> Resultado = 2 + 4
-  Si N = 3 -> Resultado = 2 + 4 + 6
-  Si N = 4 -> Resultado = 2 + 4 + 6 + 8
-  ......
-  Para cualquier N -> Resultado = 2 + 4 + 6 +....+ 2*N

````
numeroNatural = int(input("Ingrese un numero natural: "))

resultado = 0
cadena = ""

for contador in range(1, numeroNatural + 1):
    par = 2 * contador #son los numeros que se utilizaran en la cadena empieza en 1 o sea que seria 2 * (contador=1) = 2
    resultado += par

    if contador < numeroNatural:
        cadena += str(par) + " + " #esto es: la cadena empieza vacia, en el primer bucle hace 0+2=2, en el 2do bucle hace 2+2=4... hasta el numeroNatural+1
    else:
        cadena += str(par)

print(cadena + " = " + str(resultado))
````
